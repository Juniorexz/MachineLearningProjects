```python
import os
import cv2
from google.colab.patches import cv2_imshow  # Importa o método específico para exibir imagens no Colab

def process_image(image_path, binary_threshold=128):
    # Verifica se o arquivo existe
    if not os.path.exists(image_path):
        raise FileNotFoundError(f"O caminho especificado não existe: {image_path}")

    # Lê a imagem colorida
    color_image = cv2.imread(image_path)
    if color_image is None:
        raise FileNotFoundError(f"Não foi possível carregar a imagem: {image_path}")

    # Converte para tons de cinza
    gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)

    # Aplica a binarização
    _, binary_image = cv2.threshold(gray_image, binary_threshold, 255, cv2.THRESH_BINARY)

    return gray_image, binary_image

# Caminho correto para sua imagem
input_image_path = "sonic-3-shadow.jpg"

try:
    # Processa a imagem
    gray, binary = process_image(input_image_path)

    # Exibe as imagens no Colab
    print("Imagem em tons de cinza:")
    cv2_imshow(gray)

    print("Imagem binarizada:")
    cv2_imshow(binary)

except FileNotFoundError as e:
    print(e)
```
